```python
@Controlador("3", "ejercicio_pyg_11")
def ejercicio_3_11(numero: int):
    """
    Instrucciones:
        Esta función recibe un número.

        Si el número está entre 10 y 20 (incluidos),
        debe regresar el texto "dentro del rango".

        En cualquier otro caso debe regresar
        "fuera del rango".

    Ejemplo:
        Si numero = 15
        la función debe regresar "dentro del rango"

        Si numero = 25
        la función debe regresar "fuera del rango"

    Pista:
        Usa el operador lógico and.
    """

    # Código respuesta

    if numero >= 10 and numero <= 20:
        respuesta = "dentro del rango"
    else:
        respuesta = "fuera del rango"

    # Código respuesta

    return respuesta
```

```python
@Controlador("3", "ejercicio_pyg_12")
def ejercicio_3_12(edad: int, estudiante: bool):
    """
    Instrucciones:
        Esta función recibe la edad de una persona
        y un valor booleano llamado estudiante.

        Si la persona es menor de 18 o es estudiante,
        debe regresar "descuento disponible".

        En cualquier otro caso debe regresar
        "sin descuento".

    Ejemplo:
        Si edad = 16 y estudiante = False
        la función debe regresar "descuento disponible"

        Si edad = 25 y estudiante = True
        la función debe regresar "descuento disponible"

        Si edad = 30 y estudiante = False
        la función debe regresar "sin descuento"

    Pista:
        Usa el operador lógico or.
    """

    # Código respuesta

    if edad < 18 or estudiante:
        respuesta = "descuento disponible"
    else:
        respuesta = "sin descuento"

    # Código respuesta

    return respuesta
```

```python
@Controlador("3", "ejercicio_pyg_13")
def ejercicio_3_13(usuario: str, password: str):
    """
    Instrucciones:
        Esta función recibe un usuario y una contraseña.

        El acceso es válido únicamente si:
        - el usuario es "admin"
        - y la contraseña es "1234"

        Si ambas condiciones se cumplen
        debe regresar "acceso concedido".

        En cualquier otro caso debe regresar
        "acceso denegado".

    Pista:
        Usa el operador lógico and.
    """

    # Código respuesta

    if usuario == "admin" and password == "1234":
        respuesta = "acceso concedido"
    else:
        respuesta = "acceso denegado"

    # Código respuesta

    return respuesta
```

```python
@Controlador("3", "ejercicio_pyg_15")
def ejercicio_3_15(edad: int, invitacion: bool):
    """
    Instrucciones:
        Esta función recibe la edad de una persona
        y un valor booleano llamado invitacion.

        Una persona puede entrar al evento si:
        - tiene 21 años o más
        - o tiene una invitación.

        Si se cumple alguna de estas condiciones
        debe regresar "puede entrar".

        En caso contrario debe regresar
        "no puede entrar".

    Pista:
        Usa el operador lógico or.
    """

    # Código respuesta

    if edad >= 21 or invitacion:
        respuesta = "puede entrar"
    else:
        respuesta = "no puede entrar"

    # Código respuesta

    return respuesta
```
