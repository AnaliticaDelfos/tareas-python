```python
@Controlador("5", "ejercicio_pyg_21")
def ejercicio_5_21(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista de números.

        Debes usar un ciclo for para recorrer
        la lista y calcular la suma de todos
        los elementos.

    Ejemplo:
        Si lista = [1, 2, 3, 4]

        la función debe regresar 10

    Pista:
        Usa una variable acumuladora.
    """

    # Código respuesta

    suma = 0

    for numero in lista:
        suma = suma + numero

    respuesta = suma

    # Código respuesta

    return respuesta
```

```python
@Controlador("5", "ejercicio_pyg_22")
def ejercicio_5_22(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista de números.

        Debes contar cuántos números
        son mayores que 10.

    Ejemplo:
        Si lista = [5, 12, 8, 20, 3]

        la función debe regresar 2
    """

    # Código respuesta

    contador = 0

    for numero in lista:
        if numero > 10:
            contador = contador + 1

    respuesta = contador

    # Código respuesta

    return respuesta
```

```python
ejercicio_5_22(["Ana", "Luis"], [20, 25])
```

```python
calificador.califica_ejercicio_pyg_22(ejercicio_5_22)
```

```python
@Controlador("5", "ejercicio_pyg_23")
def ejercicio_5_23(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista de números.

        Debes crear una nueva lista donde cada
        número esté multiplicado por 2.

    Ejemplo:
        Si lista = [1, 2, 3]

        la función debe regresar

        [2, 4, 6]

    Pista:
        Crea una lista vacía y usa append().
    """

    # Código respuesta

    nueva_lista = []

    for numero in lista:
        nueva_lista.append(numero * 2)

    respuesta = nueva_lista

    # Código respuesta

    return respuesta
```

```python
@Controlador("5", "ejercicio_pyg_24")
def ejercicio_5_24(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista de números.

        Debes crear una nueva lista que contenga
        únicamente los números pares.

    Ejemplo:
        Si lista = [1, 2, 3, 4, 5, 6]

        la función debe regresar

        [2, 4, 6]
    """

    # Código respuesta

    pares = []

    for numero in lista:
        if numero % 2 == 0:
            pares.append(numero)

    respuesta = pares

    # Código respuesta

    return respuesta
```

```python
ejercicio_5_24([[1], [2], [3]], [2, 4, 5])
```

```python
calificador.califica_ejercicio_pyg_24(ejercicio_5_24)
```

```python
@Controlador("5", "ejercicio_pyg_25")
def ejercicio_5_25(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista de números.

        Debes recorrer la lista con un ciclo for
        y encontrar el número mayor.

    Ejemplo:
        Si lista = [3, 7, 2, 9, 5]

        la función debe regresar 9

    Pista:
        Puedes guardar el valor mayor
        en una variable y actualizarla
        cuando encuentres un número mayor.
    """

    # Código respuesta

    mayor = lista[0]

    for numero in lista:
        if numero > mayor:
            mayor = numero

    respuesta = mayor

    # Código respuesta

    return respuesta
```

```python
ejercicio_5_25([[1, 2], [2, 3], [3, 4]], [3, 6, 9], [[4, 5], [5, 6]])
```

```python
calificador.califica_ejercicio_pyg_25(ejercicio_5_25)
```
