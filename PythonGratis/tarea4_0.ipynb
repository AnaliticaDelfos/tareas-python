```python
@Controlador("4", "ejercicio_pyg_20")
def ejercicio_4_16(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista de elementos.

        Debe regresar la cantidad de elementos
        que hay dentro de la lista.

    Ejemplo:
        Si lista = [1, 2, 3, 4]

        la función debe regresar 4

    Pista:
        Puedes usar la función len().
    """

    # Código respuesta

    respuesta = len(lista)

    # Código respuesta

    return respuesta
```

```python
@Controlador("4", "ejercicio_pyg_21")
def ejercicio_4_17(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista.

        Debe regresar el primer elemento
        de la lista.

    Ejemplo:
        Si lista = [10, 20, 30]

        la función debe regresar 10

    Recuerda:
        En Python los índices empiezan en 0.
    """

    # Código respuesta

    respuesta = lista[0]

    # Código respuesta

    return respuesta
```

```python
ejercicio_4_17(["Ana", "Luis"], [20, 25])
```

```python
calificador.califica_ejercicio_pyg_17(ejercicio_4_17)
```

```python
@Controlador("4", "ejercicio_pyg_22")
def ejercicio_4_18(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista.

        Debe regresar el último elemento
        de la lista.

    Ejemplo:
        Si lista = [5, 8, 12]

        la función debe regresar 12

    Pista:
        Puedes usar índices negativos.
    """

    # Código respuesta

    respuesta = lista[-1]

    # Código respuesta

    return respuesta
```

```python
ejercicio_4_18([100, 200, 300], ["a", "b", "c"])
```

```python
calificador.califica_ejercicio_pyg_18(ejercicio_4_18)
```

```python
@Controlador("4", "ejercicio_pyg_23")
def ejercicio_4_19(lista: list, elemento):
    """
    Instrucciones:
        Esta función recibe una lista
        y un elemento.

        Debe agregar el elemento al final
        de la lista y regresar la lista
        actualizada.

    Ejemplo:
        Si lista = [1, 2, 3]
        y elemento = 4

        la función debe regresar

        [1, 2, 3, 4]

    Pista:
        Usa el método append().
    """

    # Código respuesta

    lista.append(elemento)
    respuesta = lista

    # Código respuesta

    return respuesta
```

```python
ejercicio_4_19([[1], [2], [3]], [2, 4, 5])
```

```python
calificador.califica_ejercicio_pyg_19(ejercicio_4_19)
```

```python
@Controlador("4", "ejercicio_pyg_24")
def ejercicio_4_20(lista: list):
    """
    Instrucciones:
        Esta función recibe una lista
        de números.

        Debe regresar la suma de todos
        los elementos de la lista.

    Ejemplo:
        Si lista = [1, 2, 3, 4]

        la función debe regresar 10

    Pista:
        Puedes usar la función sum().
    """

    # Código respuesta

    respuesta = sum(lista)

    # Código respuesta

    return respuesta
```

```python
ejercicio_4_20([[1, 2], [2, 3], [3, 4]], [3, 6, 9], [[4, 5], [5, 6]])
```

```python
calificador.califica_ejercicio_pyg_20(ejercicio_4_20)
```
