```python
@Controlador("2", "ejercicio_pyg_6")
def ejercicio_2_6(numero: int):
    """
    Instrucciones:
        Esta función recibe un número.

        Si el número es mayor que 0 debe regresar
        el texto "positivo".

        En cualquier otro caso debe regresar
        el texto "negativo o cero".

    Ejemplo:
        Si numero = 10
        la función debe regresar "positivo"

        Si numero = -5
        la función debe regresar "negativo o cero"
    """

    # Código respuesta

    if numero > 0:
        respuesta = "positivo"
    else:
        respuesta = "negativo o cero"

    # Código respuesta

    return respuesta
```

```python
@Controlador("2", "ejercicio_pyg_7")
def ejercicio_2_7(numero: int):
    """
    Instrucciones:
        Esta función recibe un número entero.

        Si el número es par debe regresar
        el texto "par".

        En caso contrario debe regresar
        el texto "impar".

    Ejemplo:
        Si numero = 4
        la función debe regresar "par"

        Si numero = 7
        la función debe regresar "impar"

    Pista:
        Puedes usar el operador % (módulo).
    """

    # Código respuesta

    if numero % 2 == 0:
        respuesta = "par"
    else:
        respuesta = "impar"

    # Código respuesta

    return respuesta
```

```python
ejercicio_2_7(9)
```

```python
calificador.califica_ejercicio_pyg_7(ejercicio_2_7)
```

```python
@Controlador("2", "ejercicio_pyg_8")
def ejercicio_2_8(a: int, b: int):
    """
    Instrucciones:
        Esta función recibe dos números.

        Si a es mayor que b debe regresar a.

        En cualquier otro caso debe regresar b.

    Ejemplo:
        Si a = 10
        y b = 3
        la función debe regresar 10

        Si a = 2
        y b = 8
        la función debe regresar 8
    """

    # Código respuesta

    if a > b:
        respuesta = a
    else:
        respuesta = b

    # Código respuesta

    return respuesta
```

```python
calificador.califica_ejercicio_pyg_8(ejercicio_2_8)
```

```python
@Controlador("2", "ejercicio_pyg_9")
def ejercicio_2_9(precio: float):
    """
    Instrucciones:
        Esta función recibe el precio de un producto.

        Si el precio es mayor a 100
        se debe aplicar un descuento de 10%.

        En caso contrario el precio permanece igual.
        Usa la función round para redondear a 2 decimales.

    Ejemplo:
        Si precio = 200
        la función debe regresar 180

        Si precio = 50
        la función debe regresar 50
    """

    # Código respuesta

    if precio > 100:
        respuesta = precio * 0.9
    else:
        respuesta = precio

    # Código respuesta

    return round(respuesta,2)
```

```python
calificador.califica_ejercicio_pyg_9(ejercicio_2_9)
```

```python
@Controlador("2", "ejercicio_pyg_10")
def ejercicio_2_10(edad: int):
    """
    Instrucciones:
        Esta función recibe una edad.

        Si la edad es mayor o igual a 18
        debe regresar "acceso permitido".

        En caso contrario debe regresar
        "acceso denegado".

    Ejemplo:
        Si edad = 20
        la función debe regresar "acceso permitido"

        Si edad = 16
        la función debe regresar "acceso denegado"
    """

    # Código respuesta

    if edad >= 18:
        respuesta = "acceso permitido"
    else:
        respuesta = "acceso denegado"

    # Código respuesta

    return respuesta
```
