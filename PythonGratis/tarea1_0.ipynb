```python
@Controlador("1", "ejercicio_pyg_6")
def ejercicio_1_2(a: float, b: float):
    """
    Instrucciones:
        Esta función toma dos números y devuelve
        el promedio entre ellos.
        Usa la función round para redondear a 2 decimales.

    Ejemplo:
        Si a = 6
        y b = 8
        la función debe regresar 7

    Pista:
        El promedio de dos números es la suma
        dividida entre 2.
    """

    # Código respuesta


    # Código respuesta

    return round((a + b)/2, 2) 
```

```python
@Controlador("1", "ejercicio_pyg_8")
def ejercicio_1_4(minutos: int):
    """
    Instrucciones:
        Esta función recibe una cantidad de minutos
        y devuelve el equivalente en segundos.

    Ejemplo:
        Si minutos = 2

        la función debe regresar 120

    Pista:
        Un minuto tiene 60 segundos.
    """

    # Código respuesta

    respuesta = minutos * 60

    # Código respuesta

    return respuesta
```

```python
@Controlador("1", "ejercicio_pyg_9")
def ejercicio_1_5(base: float, altura: float):
    """
    Instrucciones:
        Esta función recibe la base y la altura
        de un rectángulo y devuelve su área.
        Usa la función round para redondear a 2 decimales.

    Ejemplo:
        Si base = 5
        y altura = 3

        la función debe regresar 15

    Pista:
        El área de un rectángulo se calcula
        multiplicando base por altura.
    """

    # Código respuesta

    respuesta = base * altura

    # Código respuesta

    return round(respuesta, 2)
```
