Para las variables categóricas (como el sexo, el diagnóstico o si un paciente está hospitalizado), no podemos calcular un "promedio" con sentido clínico. Lo que nos interesa es la **Frecuencia**: ¿Cuántos hay de cada uno?

El verbo de `ggplot2` para esto es **`geom_bar()`**. A diferencia de otros gráficos, este es inteligente: él mismo cuenta cuántas veces aparece cada categoría en tu base de datos, por lo que no necesitas resumir la tabla antes de graficar.

## 1. Mi primer gráfico de barras
Vamos a ver cuál es la prevalencia de enfermedades en nuestra cohorte de 200 pacientes.

**Instrucción:** Nota que solo definimos el eje `x`. R se encarga de calcular el eje `y` (count) automáticamente.

```python
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# Graficamos la frecuencia de diagnósticos
ggplot(datos, aes(x = diagnostico)) +
  geom_bar(fill = "coral", color = "black") +
  labs(title = "Frecuencia de Diagnósticos en la Cohorte",
       x = "Patología",
       y = "Número de Pacientes") +
  theme_light()
```

## 2. Barras de colores
Podemos hacer que cada barra tenga un color distinto según la categoría usando el argumento `fill` dentro de `aes()`. Esto hace que el gráfico sea mucho más fácil de leer en una presentación.

**Instrucción:** Ejecuta la celda para ver la distribución por Sexo.

```python
ggplot(datos, aes(x = sexo, fill = sexo)) +
  geom_bar() +
  scale_fill_manual(values = c("F" = "pink", "M" = "lightblue")) +
  labs(title = "Distribución por Sexo",
       x = "Sexo",
       y = "Total")
```

## 3. Relación entre dos categorías
¿Cuántos pacientes con diabetes están hospitalizados? Podemos "apilar" información en una sola barra para ver proporciones.

**Instrucción:** Usaremos `x` para el diagnóstico y `fill` para el estado de hospitalización.

```python
ggplot(datos, aes(x = diagnostico, fill = hospitalizado)) +
  geom_bar(position = "stack") + # "stack" es el valor por defecto
  labs(title = "Hospitalización por Diagnóstico",
       x = "Diagnóstico",
       y = "Pacientes",
       fill = "¿Hospitalizado?")
```
