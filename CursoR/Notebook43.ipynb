En medicina, los fenómenos son multifactoriales. La **Regresión Lineal Múltiple** nos permite evaluar el efecto de varias variables independientes simultáneamente sobre un mismo desenlace.

### ¿Para qué sirve el Ajuste?
Imagina que observas que los pacientes con mayor edad tienen más glucosa. Pero, ¿es por la edad *per se*, o es porque los pacientes mayores también suelen tener mayor peso?

El modelo múltiple "limpia" o "ajusta" el efecto de cada variable, permitiéndonos ver el impacto real e independiente de cada una:
$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + ... + \epsilon$$

**Interpretación:** El coeficiente de la Edad ($\beta_1$) representa cuánto cambia la glucosa por cada año de vida, **manteniendo constantes** las demás variables (como el peso).

## 1. Agregando Predictores
Para agregar más variables al modelo en R, simplemente usamos el signo `+` dentro de la fórmula de la función `lm()`.

Analizaremos la **Glucosa** como desenlace, usando como predictores la **Edad** y el **Peso**.

```r
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# Modelo múltiple: Glucosa explicada por Edad Y Peso
modelo_multiple <- lm(glucosa_mgdl ~ edad + peso_kg, data = datos)

# Resumen del modelo
summary(modelo_multiple)
```

## 2. Significancia Independiente
Al ver los resultados, ahora tenemos una fila para cada predictor.
* Si el valor **p** de la Edad sigue siendo < 0.05 después de incluir el Peso, decimos que la Edad es un **predictor independiente** de la glucosa.

```r
# Visualización rápida de los coeficientes (excluyendo el intercepto)
coeficientes <- broom::tidy(modelo_multiple) |> filter(term != "(Intercept)")

ggplot(coeficientes, aes(x = term, y = estimate, fill = term)) +
  geom_col() +
  geom_errorbar(aes(ymin = estimate - std.error, ymax = estimate + std.error), width = 0.2) +
  labs(title = "Efecto Estimado de cada Variable en la Glucosa",
       subtitle = "Los bigotes representan el error estándar",
       x = "Predictor", y = "Efecto (Beta)") +
  theme_minimal()
```
