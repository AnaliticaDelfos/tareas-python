## 1. Primeras y últimas impresiones
A veces solo queremos ver una "rebanada" de la tabla.
* `head()`: Nos muestra las primeras 6 filas.
* `tail()`: Nos muestra las últimas 6 filas.

**Instrucción:** Vamos a usar la base `datos_pacientes` que importamos en el cuaderno anterior. Ejecuta la celda:

```python
# Primero nos aseguramos de tener los datos (recuerda que el archivo debe estar cargado)
library(readr)
datos_pacientes <- read_csv("cohorte_estudio.csv")

# Vemos el inicio de la tabla
head(datos_pacientes)

# Vemos el final de la tabla
tail(datos_pacientes)
```

## 2. Radiografía de los datos: `str()` y `glimpse()`
Estas funciones nos dicen cómo está construida la tabla por dentro. Nos muestran el nombre de cada columna y qué tipo de dato contiene (`num` para números, `chr` para texto, etc.).

* `str()`: Es la función clásica de R.
* `glimpse()`: Es una versión más moderna y fácil de leer (requiere la librería `dplyr`).

```python
# Radiografía clásica
str(datos_pacientes)

# Radiografía moderna (activamos dplyr primero)
library(dplyr)
glimpse(datos_pacientes)
```

## 3. Dimensiones: ¿Cuántos somos?
Para obtener el conteo exacto de filas (pacientes) y columnas (variables), usamos `dim()`.

**Instrucción:** Ejecuta la celda para obtener el tamaño exacto.

```python
dim(datos_pacientes)
```
