Para esto usaremos `ggplot2`, la herramienta más potente de R para crear gráficos de alta calidad.

## 1. ¿Cómo funciona ggplot2?
Un gráfico en R se construye por capas:
1.  **`ggplot(data = ...)`**: Le decimos qué tabla usar.
2.  **`aes(x = ...)`**: Definimos la "estética" (qué columna va en el eje X).
3.  **`geom_histogram()`**: Definimos la "geometría" (qué tipo de gráfico queremos).

```r
library(tidyverse)

# Cargamos la base de datos limpia para análisis
datos <- read_csv("cohorte_analisis.csv")

# Creamos nuestro primer histograma de Edad
ggplot(datos, aes(x = edad)) +
  geom_histogram(fill = "steelblue", color = "white") +
  labs(title = "Distribución de Edad en la Cohorte",
       x = "Edad (años)",
       y = "Frecuencia")
```

## 2. El ancho de las barras (Bins)
El aspecto de un histograma cambia según el número de barras (`bins`). Si pones muy pocas, pierdes detalle; si pones muchas, el gráfico se vuelve "ruidoso".

**Instrucción:** Prueba cambiar el valor de `bins` en el código de abajo (intenta con 10, 30 y 50) y observa cómo cambia la "forma" de la distribución.

```r
ggplot(datos, aes(x = glucosa_mgdl)) +
  geom_histogram(bins = 30, fill = "darkgreen", color = "white") +
  labs(title = "Distribución de Glucosa",
       subtitle = "¿Se parece a una campana de Gauss?",
       x = "Glucosa (mg/dL)",
       y = "Número de Pacientes")
```
