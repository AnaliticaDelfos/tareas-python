# CS106 - Sesión 3 | Notebook 14: Jerarquizando la información (arrange) 📋

Una base de datos desordenada es difícil de leer. El verbo **`arrange()`** nos permite ordenar nuestras filas (pacientes) con base en el valor de una o más columnas.

Esto es especialmente útil para:
* Ver los valores más altos o más bajos de un laboratorio.
* Organizar a los pacientes por orden alfabético.
* Ordenar cronológicamente por fecha de ingreso.

## 1. Ordenar de menor a mayor
Por defecto, `arrange()` ordena los datos de forma ascendente (de la A a la Z, o del número más pequeño al más grande).

**Instrucción:** Vamos a ordenar nuestra cohorte por edad, para ver a los pacientes más jóvenes primero.

```python
library(tidyverse)
datos <- read_csv("cohorte_estudio.csv")

# Ordenamos por edad
datos_jovenes <- datos |>
  arrange(edad)

# Vemos el resultado
datos_jovenes
```

## 2. Ordenar de mayor a menor (Prioridad)
En urgencias o cuidados críticos, nos interesa lo opuesto: ver los valores más altos primero. Para esto, envolvemos el nombre de la columna en la función `desc()`.

**Instrucción:** Ordenemos la tabla para ver quién tiene los niveles de glucosa más elevados.

```python
# Usamos desc() para orden descendente
datos_criticos <- datos |>
  arrange(desc(glucosa_mgdl))

datos_criticos
```

## 3. Orden Alfabético
`arrange()` también funciona con texto. Es muy útil para organizar listas de nombres o diagnósticos.

**Instrucción:** Ordena la tabla alfabéticamente por el nombre del paciente.

```python
datos |>
  arrange(nombre)
```
