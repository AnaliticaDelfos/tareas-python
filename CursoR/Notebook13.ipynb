# CS106 - Sesión 3 | Notebook 13: Enfocando en subgrupos (filter) 🔍

En bioestadística, rara vez analizamos a todos los pacientes al mismo tiempo. Lo normal es que necesitemos aislar subgrupos: "solo mujeres", "solo mayores de 60 años" o "solo pacientes con diabetes".

Para esto usamos el verbo **`filter()`**.

### Los Operadores Lógicos
Para filtrar, necesitamos usar comparaciones:
* `==` : Igual a (¡Nota que son dos signos de igual!).
* `!=` : Diferente de.
* `>` : Mayor que.
* `<` : Menor que.
* `>=` o `<=` : Mayor/Menor o igual que.

## 1. Filtrar por texto (Categorías)
Si queremos buscar un diagnóstico específico, usamos el operador `==` y ponemos el texto entre comillas.

**Instrucción:** Vamos a buscar solo a los pacientes que tienen "Diabetes".

```python
library(tidyverse)
datos <- read_csv("cohorte_estudio.csv")

# Filtramos la tabla
pacientes_diabetes <- datos |>
  filter(diagnostico == "Diabetes")

# Vemos el resultado
pacientes_diabetes
```

## 2. Filtrar por valores numéricos
Podemos usar comparaciones matemáticas para encontrar pacientes en rangos específicos de edad o laboratorio.

**Instrucción:** Busquemos a los pacientes mayores de 50 años.

```python
datos |>
  filter(edad > 50)
```

## 3. Combinando criterios
¿Qué pasa si busco un subgrupo muy específico? Podemos separar los criterios con una coma (que funciona como un "Y").

**Instrucción:** Busquemos hombres (`sexo == "M"`) que además estén hospitalizados (`hospitalizado == TRUE`).

```python
datos |>
  filter(sexo == "M", hospitalizado == TRUE)
```
