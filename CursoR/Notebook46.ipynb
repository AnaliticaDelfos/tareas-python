En medicina, muchas veces el desenlace no es un "sí" o un "no" estático, sino un proceso que ocurre en el tiempo. El **Análisis de Supervivencia** se utiliza para estudiar el tiempo que transcurre hasta que ocurre un evento de interés (muerte, alta hospitalaria, recaída, etc.).

### Conceptos Clave:
1. **Tiempo al Evento:** El intervalo (días, meses, años) desde un punto de partida definido hasta el suceso.
2. **Evento (Status):** Variable binaria (1 = Ocurrió el evento, 0 = No ha ocurrido aún).
3. **Censura:** Ocurre cuando el estudio termina o el paciente se retira sin haber presentado el evento. No sabemos cuándo ocurrirá, pero sabemos que hasta ese momento "sobrevivió".



### El Objeto `Surv`:
En R, para analizar supervivencia, combinamos el tiempo y el estado en un solo objeto especial usando la librería `survival`.

## 1. Identificación de Variables Clínicas
Para este análisis, utilizaremos dos columnas fundamentales:
* **`dias_hasta_evento`**: El tiempo de seguimiento.
* **`evento`**: Donde `1` representa el suceso (ej. reingreso) y `0` representa censura (el paciente sigue bien al final del estudio).

```python
library(tidyverse)
library(survival)

# Cargamos los datos reales
datos <- read_csv("cohorte_analisis.csv")

# Creamos el objeto de supervivencia Surv(tiempo, evento)
# Nota: Observa cómo los valores censurados aparecen con un signo '+'
objeto_surv <- Surv(time = datos$dias_hasta_evento, event = datos$evento)

# Visualizamos los primeros 10 registros del objeto Surv
head(objeto_surv, 10)
```

## 2. ¿Qué tan rápido ocurre el evento?
Antes de hacer curvas complejas, calculamos cuántos eventos ocurrieron en relación con el tiempo total de seguimiento de todos los pacientes (persona-tiempo). Esto nos da la **Tasa de Incidencia**.

```python
total_eventos <- sum(datos$evento)
tiempo_total <- sum(datos$dias_hasta_evento)

tasa_incidencia <- (total_eventos / tiempo_total) * 1000

cat("Total de Eventos:", total_eventos, "\n")
cat("Tiempo Total de Seguimiento (días):", tiempo_total, "\n")
cat("Tasa de Incidencia (por 1,000 personas-día):", round(tasa_incidencia, 2))
```
