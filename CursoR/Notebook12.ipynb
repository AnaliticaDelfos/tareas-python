# CS106 - Sesión 3 | Notebook 12: Eligiendo mis variables (select) ✂️

En investigación, trabajar con tablas demasiado anchas distrae y consume memoria. El verbo `select()` nos permite quedarnos únicamente con las columnas que nos interesan para nuestro análisis actual.

Podemos seleccionar columnas de varias formas:
1. Por **nombre** exacto.
2. Por **posición** (ej. la primera y la tercera).
3. Por **patrones** (ej. "todas las que empiecen con la palabra glucosa").

## 1. Selección básica por nombre
Esta es la forma más común. Simplemente listamos los nombres de las columnas que queremos conservar.

**Instrucción:** Vamos a cargar nuestra base y seleccionar solo el nombre y el diagnóstico de los pacientes.

```python
library(tidyverse)
datos <- read_csv("cohorte_estudio.csv")

# Usamos el Pipe para pasar los datos al verbo select
datos_resumen <- datos |>
  select(nombre, diagnostico)

# Vemos el resultado (la tabla ahora es más delgada)
datos_resumen
```

## 2. Quitar lo que no sirve
A veces es más fácil decir qué columna **NO** queremos. Para eso usamos el signo de resta `-`.

**Instrucción:** Ejecuta la celda para ver la tabla SIN la columna `id_paciente`.

```python
datos |>
  select(-id_paciente)
```

## 3. Selecciones inteligentes
Dplyr tiene funciones que nos ayudan a seleccionar grupos de columnas rápidamente:
* `starts_with("abc")`: Columnas que empiezan con...
* `ends_with("xyz")`: Columnas que terminan con...
* `contains("123")`: Columnas que contienen...

**Instrucción:** Seleccionemos todas las variables relacionadas con la glucosa.

```python
datos |>
  select(nombre, starts_with("glucosa"))
```
