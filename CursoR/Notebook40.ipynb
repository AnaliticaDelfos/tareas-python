## Escenario A: Estatura según Sexo
¿Existe una diferencia significativa en la estatura entre hombres y mujeres? Primero verificaremos la normalidad de la estatura y luego aplicaremos la prueba correspondiente.

```r
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# 1. Verificación de Normalidad (Shapiro-Wilk)
print(shapiro.test(datos$estatura_m))

# 2. Comparación de Medias (t-Student)
test_a <- t.test(estatura_m ~ sexo, data = datos)
print(test_a)
```

## Escenario B: Glucosa según Hospitalización
Analizaremos si los niveles de glucosa varían en pacientes hospitalizados. Dado que la glucosa suele presentar valores extremos, utilizaremos una prueba robusta.

```r
# Comparación de rangos (Wilcoxon)
test_b <- wilcox.test(glucosa_mgdl ~ hospitalizado, data = datos)
medianas_b <- datos |> group_by(hospitalizado) |> summarise(mediana = median(glucosa_mgdl))

print(test_b)
print(medianas_b)
```

## Escenario C: Diagnóstico vs. Hospitalización
¿Existe una asociación entre el tipo de diagnóstico (Diabetes, Hipertensión, etc.) y la necesidad de hospitalización?

```r
# Tabla de contingencia y Chi-cuadrada
tabla_c <- table(datos$diagnostico, datos$hospitalizado)
test_c <- chisq.test(tabla_c)

print(tabla_c)
print(test_c)
```
