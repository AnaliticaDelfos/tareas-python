# CS106 - Sesión 2 | Notebook 09: El misterio de los datos faltantes (NA) 🔍

En la práctica clínica, es muy común que a un paciente no se le tome una muestra, que pierda una cita o que simplemente el dato no se registre.

En **R**, estos "huecos" se representan con la etiqueta **`NA`** (del inglés *Not Available* / No Disponible). Es vital aprender a detectarlos antes de hacer cualquier cálculo estadístico.

## 1. Identificando el vacío: `is.na()`
La función `is.na()` le pregunta a R: "¿Este dato está vacío?".
* Si la celda está vacía, R responde `TRUE`.
* Si tiene un dato, R responde `FALSE`.

**Instrucción:** Vamos a revisar nuestra base de datos.

```python
library(readr)
datos_pacientes <- read_csv("cohorte_estudio.csv")

# ¿Cuáles celdas son NA? (Aparecerá una tabla de TRUE/FALSE)
is.na(datos_pacientes)
```

## 2. Contando los "huecos"
Ver una tabla de `TRUE/FALSE` es difícil si tenemos muchos pacientes. Es mejor pedirle a R que sume cuántos `NA` hay en total.

**Instrucción:** Ejecuta la celda para ver el conteo total de datos faltantes en toda la base.

```python
# Sumamos todos los TRUE que encontró is.na
sum(is.na(datos_pacientes))
```

## 3. La visión global: `summary()`
La función `summary()` es nuestra mejor amiga para el diagnóstico inicial. Nos da el promedio, el mínimo, el máximo y, lo más importante, **nos avisa si hay NAs** en cada columna.

**Instrucción:** Ejecuta la celda y busca la etiqueta `NA's` al final de cada columna.

```python
summary(datos_pacientes)
```
