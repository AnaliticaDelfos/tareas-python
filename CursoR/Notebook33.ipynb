# CS106 - Sesión 7 | Notebook 33: Comparación Grupal (`by =`) ⚖️

El análisis de una cohorte cobra sentido cuando comparamos subgrupos. El argumento principal para esto en `gtsummary` es **`by`**.

Al usar `by`, R divide automáticamente tu tabla en columnas, calculando los estadísticos para cada grupo por separado. Hoy aprenderemos a:
1. Segmentar la tabla por variables categóricas.
2. Añadir una columna de **Total** (`add_overall()`) para comparar los grupos contra el promedio general.
3. Añadir el conteo de muestras por fila (`add_n()`).

## 1. El argumento `by`
Imagina que quieres ver si los niveles de glucosa o la edad varían según el sexo de los pacientes.

**Instrucción:** Vamos a crear una tabla comparativa básica usando `by = sexo`.

```r
library(tidyverse)
library(gtsummary)
library(gt)
library(IRdisplay)

datos <- read_csv("cohorte_analisis.csv")
```

```r

tabla_html <- datos |>
  select(sexo, edad, glucosa_mgdl, peso_kg) |>
  tbl_summary(
    by = sexo,
    label = list(
      edad ~ "Edad",
      glucosa_mgdl ~ "Glucosa",
      peso_kg ~ "Peso"
    )
  ) |>
  bold_labels() |>
  as_gt() |>
  gt::as_raw_html()

IRdisplay::display_html(tabla_html)
```

## 2. Añadiendo la columna "Overall"
A veces, al dividir por grupos, perdemos de vista cómo se comporta la población completa. La función `add_overall()` soluciona esto agregando una columna con el resumen de todos los pacientes.

**Instrucción:** Ejecuta la celda y observa cómo aparece la columna "Total".

```r
tabla_html <- datos |>
  select(diagnostico, edad, glucosa_mgdl, hospitalizado) |>
  tbl_summary(by = hospitalizado) |>
  add_overall(
    last = FALSE,
    col_label = "**Total (N = {N})**"
  ) |>
  bold_labels() |>
  as_gt() |>
  gt::as_raw_html()

IRdisplay::display_html(tabla_html)
```

## 3. ¿Cuántos datos tenemos realmente?
En tablas grandes, es útil saber cuántos pacientes contribuyeron a cada fila (especialmente si hay datos faltantes, aunque aquí nuestra base está limpia). La función `add_n()` agrega esta transparencia.

**Instrucción:** Agregaremos el conteo `n` a nuestra tabla de diagnósticos.

```r
tabla_html <- datos |>
  select(diagnostico, edad, peso_kg, estatura_m) |>
  tbl_summary(by = diagnostico) |>
  add_n() |>
  modify_header(n = "**N total**") |>
  bold_labels() |>
  as_gt() |>
  gt::as_raw_html()

IRdisplay::display_html(tabla_html)
```
