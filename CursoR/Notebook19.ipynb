# CS106 - Sesión 4 | Notebook 19: El estetoscopio estadístico (summarise) 🩺

Si `mutate()` servía para añadir columnas, **`summarise()`** sirve para "colapsar" la tabla. Toma cientos de filas y las resume en un solo valor estadístico (un promedio, una desviación estándar, un valor máximo, etc.).

Es como usar un estetoscopio para escuchar el corazón de toda una población: no nos interesa el latido individual, sino el ritmo general del grupo.

## 1. Resumen de toda la base
Podemos obtener métricas generales de toda nuestra cohorte. Las funciones más usadas son:
* `mean()`: Promedio.
* `sd()`: Desviación estándar.
* `n()`: Conteo total.

**Instrucción:** Vamos a calcular la edad promedio de toda nuestra base de datos.

```python
library(tidyverse)
datos <- read_csv("cohorte_estudio.csv")

# Resumen general
datos |>
  summarise(
    edad_promedio = mean(edad),
    total_pacientes = n()
  )
```

## 2. El análisis por grupos
Aquí es donde ocurre la magia. Si primero usamos `group_by()` y luego `summarise()`, R calculará las estadísticas por separado para cada categoría. **Este es el formato estándar de una Tabla 1 en un artículo médico.**

**Instrucción:** Comparemos la glucosa promedio entre hombres y mujeres.

```python
# Primero agrupamos por sexo, luego resumimos
datos |>
  group_by(sexo) |>
  summarise(
    glucosa_promedio = mean(glucosa_mgdl),
    desviacion_estandar = sd(glucosa_mgdl)
  )
```

## 3. La "Foto Completa"
Puedes pedirle a R que te dé un perfil completo de una variable en una sola tabla de resumen.

**Instrucción:** Vamos a ver el mínimo, el máximo y el promedio de glucosa según el diagnóstico.

```python
datos |>
  group_by(diagnostico) |>
  summarise(
    minima = min(glucosa_mgdl),
    promedio = mean(glucosa_mgdl),
    maxima = max(glucosa_mgdl),
    conteo = n()
  )
```
