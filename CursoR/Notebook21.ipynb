Antes de decidir cómo arreglar una base de datos, debemos saber qué tan "rota" está. En bioestadística, no solo nos importa **cuántos** datos faltan, sino **dónde** faltan y si hay un patrón oculto.

Hoy utilizaremos herramientas visuales para crear un "mapa de calor" de los datos faltantes. Esto nos permite identificar si la pérdida de información es aleatoria o si hay variables que sistemáticamente no se recolectaron.

## 1. Activando el radar de NAs
Para esta sesión, además del `tidyverse`, utilizaremos una librería llamada `naniar`. Esta librería está diseñada específicamente para el análisis de datos faltantes en investigación.

**Instrucción:** Ejecuta la celda para cargar las librerías.

```r
library(tidyverse)
# Instalamos naniar si no está (solo la primera vez)
# install.packages("naniar")
library(naniar)

# Cargamos nuestra nueva base de datos de 200 pacientes
datos <- read_csv("cohorte_limpieza.csv")
```

## 2. ¿Cuántos huecos tenemos?
Antes de ir a lo visual, veamos un resumen rápido por columna. La función `miss_var_summary()` nos da un reporte detallado del conteo y el porcentaje de pérdida por cada variable.

**Instrucción:** Ejecuta la celda y observa qué variables tienen mayor porcentaje de pérdida.

```r
# Reporte detallado de datos faltantes
datos |>
  miss_var_summary()
```

## 3. Visualizando el Mapa del Vacío
A veces, los números no cuentan toda la historia. La función `vis_miss()` crea un gráfico donde cada raya negra representa un dato faltante. Esto nos ayuda a ver si hay pacientes que tienen "bloques" enteros de información perdida.

**Instrucción:** Ejecuta la celda para generar el mapa de calor.

```r
# Generamos el mapa visual de NAs
vis_miss(datos) +
  labs(title = "Mapa de Datos Faltantes: Cohorte de Limpieza")
```
