¿Las diferencias que observamos entre grupos son producto del azar o tienen significancia estadística? Para responder esto, necesitamos el **valor p**.

En este cuaderno aprenderemos la función más potente de `gtsummary`: **`add_p()`**. Esta función es capaz de:
1. Identificar si la variable es continua o categórica.
2. Seleccionar la prueba estadística adecuada (T-test, Wilcoxon, Chi-cuadrada o Fisher).
3. Ejecutar la prueba y añadir el resultado a nuestra Tabla 1 en una sola columna.

## 1. Implementación básica
Cuando añadimos `add_p()` después de `tbl_summary(by = ...)`, R realiza las comparaciones entre las columnas automáticamente.

**Instrucción:** Vamos a comparar las variables clínicas según el sexo de los pacientes. Observa las notas al pie de la tabla final; ahí R te dirá qué prueba usó para cada fila.

```python
library(tidyverse)
library(gtsummary)
library(gt)
library(IRdisplay)

datos <- read_csv("cohorte_analisis.csv")
```

```python

tabla_html <- datos |>
  select(sexo, edad, glucosa_mgdl, peso_kg, diagnostico) |>
  tbl_summary(by = sexo) |>
  add_p() |> # Agrega la columna de p-value
  bold_labels() |>
  as_gt() |>
  gt::as_raw_html()

IRdisplay::display_html(tabla_html)
```

## 2. Enfatizando resultados: `bold_p()`
En un artículo, queremos que los resultados con significancia estadística ($p < 0.05$) salten a la vista. La función `bold_p()` pone en negrita automáticamente esos valores.

También podemos usar `add_q()` si quisiéramos ajustar por comparaciones múltiples (aunque eso es un tema más avanzado).

```python
tabla_html <- datos |>
  select(hospitalizado, edad, glucosa_mgdl, peso_kg, diagnostico) |>
  tbl_summary(by = hospitalizado) |>
  add_p() |>
  bold_p(t = 0.05) |> # Resalta con p < 0.05
  add_overall() |>
  bold_labels() |>
  as_gt() |>
  gt::as_raw_html()

IRdisplay::display_html(tabla_html)
```

## 3. El algoritmo de gtsummary
Es vital que como investigadores sepamos qué está pasando "bajo el capó". Por defecto:
* **Variables Continuas:** Usa la prueba de **Suma de Rangos de Wilcoxon** (No paramétrica).
* **Variables Categóricas:** Usa **Chi-cuadrada** de Pearson o **Prueba Exacta de Fisher** si hay pocos datos.

Si tú quieres forzar una prueba específica (ej. un T-test porque sabes que tus datos son normales), puedes hacerlo dentro de `add_p()`.

```python
# Forzamos un T-test para la edad y el peso
tabla_html <- datos |>
  select(hospitalizado, edad, peso_kg) |>
  tbl_summary(by = hospitalizado) |>
  add_p(
    test = list(
      all_continuous() ~ "t.test"
    )
  ) |>
  bold_labels() |>
  as_gt() |>
  gt::as_raw_html()

IRdisplay::display_html(tabla_html)
```
