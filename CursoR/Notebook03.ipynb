### ¿Qué es una Variable?
Imagina que una variable es una **caja con una etiqueta**.
1. Guardas un dato dentro de la caja (un peso, una edad, un nombre).
2. Le pones un nombre a la etiqueta.
3. Cada vez que menciones ese nombre, R sabrá exactamente qué hay dentro.

### El Operador de Asignación `<-`
Para guardar datos en R, usamos una "flecha" llamada operador de asignación: `<-`. Se lee como "guarda esto en...".

## 3. La "Magia" de las Variables: El IMC Reutilizable
¿Recuerdas la fórmula del IMC? `peso / (estatura * estatura)`.

Al usar variables, el código se vuelve mucho más fácil de leer y entender, casi como una receta médica.

**Instrucción:** Ejecuta esta celda donde definimos los datos del paciente y calculamos su resultado.

```python
# 1. Definimos los datos del paciente
peso_kg <- 75
estatura_m <- 1.72

# 2. Realizamos el cálculo usando los nombres de las cajas
imc_resultado <- peso_kg / (estatura_m * estatura_m)

# 3. Vemos el resultado final
print(imc_resultado)
```

## ⚠️ Reglas de Oro para las etiquetas (Variables)
Para que R no se confunda, debemos seguir estas reglas:
1. **Sin espacios:** Usa guiones bajos (ej: `peso_paciente` en lugar de `peso paciente`).
2. **Sin empezar con números:** El nombre debe empezar con una letra.
3. **Minúsculas preferentemente:** R es sensible (no es lo mismo `Peso` que `peso`).
