# CS106 - Sesión 7 | Notebook 31: Mi primera tabla descriptiva (`gtsummary`) 📝

En investigación clínica, la **"Tabla 1"** es el estándar de oro. Es la primera tabla de cualquier artículo científico y su función es describir las características basales de la población de estudio (edad, sexo, diagnósticos, etc.).

Tradicionalmente, construir esta tabla implica cruzar datos, contar frecuencias y calcular promedios manualmente. Hoy aprenderemos a usar el paquete **`gtsummary`**, que permite generar una tabla con calidad de publicación de forma automática y profesional.

## 1. Carga de Librerías y Datos
Para esta sesión utilizaremos `tidyverse` para la manipulación y el paquete especializado `gtsummary` para la creación de tablas.

**Instrucción:** Ejecuta la celda para cargar las herramientas y nuestra base de datos limpia (`cohorte_analisis.csv`).

```r
library(tidyverse)
# install.packages("gtsummary") # Solo si no está instalado en el entorno
library(gtsummary)

# Cargamos la base de datos analítica
datos <- read_csv("cohorte_analisis.csv")
```

## 2. Selección inteligente de variables
En una base de datos real con 50 o 100 columnas, no querrás seleccionarlas una por una. `gtsummary` se lleva muy bien con los "selectores" de Tidyverse.

**Ejercicio 2.1: Resumen por tipo de dato**
Imagina que solo quieres ver las métricas físicas (numéricas). Podemos usar `where(is.numeric)` para que R las encuentre por nosotros.

```r
# Resumir solo variables numéricas (saltando el ID)
datos |>
  select(-id_paciente) |> # Quitamos el ID porque no es una métrica clínica
  select(where(is.numeric)) |>
  tbl_summary()
```

## 3. ¿Qué resumir y qué no?
No todas las variables aportan valor en una Tabla 1. El ID del paciente o el nombre son datos de identificación, no estadísticos.
