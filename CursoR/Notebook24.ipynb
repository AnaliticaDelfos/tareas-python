# CS106 - Sesión 5 | Notebook 24: El dilema de borrar (na.omit vs. drop_na) 🗑️

No todos los datos pueden o deben ser rescatados. Si a un paciente le faltan sus tres variables principales (ej. edad, diagnóstico y glucosa), imputar todo podría "inventar" un paciente que no existe realmente.

En este cuaderno aprenderemos a realizar una **limpieza quirúrgica**:
1.  **`na.omit()`**: El método "nuclear" (borra cualquier fila que tenga al menos un NA).
2.  **`drop_na()`**: El método "quirúrgico" (borra filas solo si faltan variables críticas).
3.  **Impacto en la muestra ($n$):** Siempre debemos vigilar cuántos pacientes perdemos en el proceso.

## 1. El riesgo de `na.omit()`
Esta función es sencilla pero peligrosa. Elimina a cualquier paciente que tenga un solo hueco en cualquier columna. En bases de datos grandes, esto puede reducir tu muestra a la mitad.

**Instrucción:** Vamos a ver cuántos pacientes nos quedan si usamos este método.

```r
library(tidyverse)
datos <- read_csv("cohorte_limpieza.csv")

# Aplicamos la limpieza total
datos_nuclear <- datos |>
  na.omit()

# Comparamos el tamaño original vs el nuevo
print(paste("Original:", nrow(datos), "pacientes"))
print(paste("Con na.omit:", nrow(datos_nuclear), "pacientes"))
```

## 2. Eliminación dirigida con `drop_na()`
Es mucho mejor decir: "Solo elimina al paciente si le falta la **Glucosa**, porque es mi variable principal". Esto protege el resto de tus datos.

**Instrucción:** Elimina solo a los pacientes que no tengan dato de glucosa.

```r
# Borramos solo si falta glucosa_mgdl
datos_filtrados <- datos |>
  drop_na(glucosa_mgdl)

# Observa que conservamos más pacientes que con el método anterior
nrow(datos_filtrados)
```

## 3. Filtrado por integridad clínica
A veces no borramos por `NA`, sino por coherencia. Por ejemplo, si tu estudio es solo para adultos, debes eliminar a cualquier menor de 18 años que se haya colado en la base.

**Instrucción:** Filtra la base para quedarte solo con mayores de 18 años.

```r
datos_adultos <- datos |>
  filter(edad >= 18)

nrow(datos_adultos)
```
