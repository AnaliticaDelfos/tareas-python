# CS106 - Sesión 4 | Notebook 16: El alquimista de datos (mutate) 🧪

En bioestadística, muchas veces los datos crudos no son suficientes. Necesitamos transformarlos: convertir libras a kilos, calcular un índice o crear una escala de riesgo.

El verbo **`mutate()`** (mutar) nos permite crear nuevas columnas en nuestra tabla sin borrar las anteriores. Es como añadir una columna nueva al final de tu hoja de Excel, pero de forma automática para miles de pacientes.

## 1. Operaciones básicas en toda la columna
Imagina que recibiste los niveles de glucosa en mg/dL, pero tu protocolo requiere **mmol/L**. La fórmula es dividir entre 18.01.

**Instrucción:** Vamos a cargar nuestra base y crear esa nueva columna.

```python
library(tidyverse)
datos <- read_csv("cohorte_estudio.csv")

# Creamos la nueva columna 'glucosa_mmol'
datos_transformados <- datos |>
  mutate(glucosa_mmol = glucosa_mgdl / 18.01)

# Observa que ahora hay una columna extra al final
datos_transformados |> select(nombre, glucosa_mgdl, glucosa_mmol)
```

## 2. El Indicador Estrella: IMC
Como vimos en la primera sesión, el IMC es `peso / (estatura * estatura)`. Con `mutate()`, podemos calcularlo para toda nuestra cohorte en un solo paso.

**Instrucción:** Ejecuta la celda para calcular el IMC de todos los pacientes.

```python
# Calculamos el IMC para todos
datos_con_imc <- datos |>
  mutate(imc = peso_kg / (estatura_m * estatura_m))

# Verificamos los resultados
datos_con_imc |> select(nombre, peso_kg, estatura_m, imc)
```

## 3. Varias columnas a la vez
Puedes crear tantas columnas como necesites dentro de un mismo `mutate()`, separándolas por comas.

**Instrucción:** Vamos a crear una columna con la edad en meses y otra con el peso en gramos.

```python
datos |>
  mutate(
    edad_meses = edad * 12,
    peso_gr = peso_kg * 1000
  ) |>
  select(nombre, edad_meses, peso_gr)
```
