En **R**, a estas tablas las llamamos **Data Frames**.

### La Anatomía de un Data Frame
Un Data Frame es muy parecido a una hoja de Excel:
* **Filas:** Representan las **Observaciones** (en nuestro caso, suele ser cada paciente).
* **Columnas:** Representan las **Variables** (los parámetros que medimos: edad, peso, diagnóstico).
