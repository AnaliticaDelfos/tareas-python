# CS106 - Sesión 4 | Notebook 18: El "Sombrero Seleccionador" (group_by) 🎩

En investigación clínica, casi siempre queremos comparar grupos: ¿Cómo le fue a los hombres frente a las mujeres? ¿Hay diferencia entre el grupo de control y el de tratamiento?

El verbo **`group_by()`** funciona como el "Sombrero Seleccionador". No cambia los datos, pero le pone una "etiqueta invisible" a cada fila para que R sepa a qué grupo pertenece.

**Nota importante:** Cuando ejecutas `group_by()`, la tabla se verá igual, pero en el encabezado verás una línea que dice `Groups:`. ¡Esa es la señal de que funcionó!

## 1. Agrupando por una variable categórica
Vamos a pedirle a R que agrupe a nuestros pacientes por su estado de hospitalización.

**Instrucción:** Ejecuta la celda y observa el encabezado del resultado.

```python
library(tidyverse)
datos <- read_csv("cohorte_estudio.csv")

# Agrupamos por hospitalización
datos_agrupados <- datos |>
  group_by(hospitalizado)

# Al imprimirlo, busca la línea que dice "Groups: hospitalizado [2]"
datos_agrupados
```

## 2. ¿Para qué sirve agrupar?
Por sí solo, `group_by()` parece no hacer mucho. Sin embargo, es el motor que permite que el siguiente verbo (`summarise`) funcione.

Imagina que quieres contar cuántos pacientes hay en cada diagnóstico. Si agrupas primero, R hará el conteo por separado para "Diabetes", "Hipertensión", etc.

```python
# Usamos n() para contar cuántos hay en cada grupo
datos |>
  group_by(diagnostico) |>
  summarise(total = n())
```
