Cuando trabajamos con variables cualitativas (texto), la lógica de la media o la mediana no aplica. Si a un paciente le falta el diagnóstico, tenemos dos caminos principales:
1.  **Etiqueta de control:** Asignar un valor constante como "No registrado" o "Desconocido". Esto es muy útil para mantener la transparencia del estudio.
2.  **Imputación por la Moda:** Asignar la categoría que más se repite en la base de datos (ej. si el 80% son hombres, asumimos que el dato faltante podría ser hombre, aunque esto requiere precaución clínica).

Hoy aprenderemos a estandarizar nuestras categorías para que los modelos estadísticos no ignoren a estos pacientes.

## 1. El valor "No registrado"
Esta es la técnica más segura en investigación médica. En lugar de inventar un dato, etiquetamos el vacío para que sea una categoría más en nuestro análisis.

**Instrucción:** Vamos a cargar la base y llenar los diagnósticos vacíos con la frase "Sin diagnóstico".

```r
library(tidyverse)
datos <- read_csv("cohorte_limpieza.csv")

# Imputamos la variable diagnóstico
datos_etiquetados <- datos |>
  mutate(diagnostico = replace_na(diagnostico, "Sin diagnóstico"))

# Comprobamos los niveles de la variable
datos_etiquetados |>
  group_by(diagnostico) |>
  summarise(conteo = n())
```

## 2. Llenando con el valor más frecuente (Moda)
A veces, para variables como el `sexo`, se prefiere usar la moda. R no tiene una función "mode" directa para texto, pero podemos identificarla fácilmente con un conteo.

**Instrucción:** Identifica cuál es el sexo más frecuente y úsalo para llenar los huecos.

```r
# 1. Identificar la moda
datos |> count(sexo, sort = TRUE)

# 2. Supongamos que la moda es "M". Imputamos:
datos_sexo_listo <- datos |>
  mutate(sexo = replace_na(sexo, "M"))

# Verificamos
sum(is.na(datos_sexo_listo$sexo))
```
