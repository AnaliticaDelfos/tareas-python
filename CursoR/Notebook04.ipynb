# CS106 - Sesión 1 | Notebook 04: Los tipos de datos en la salud 🩺

No toda la información en un expediente clínico es igual. R clasifica los datos en diferentes "tipos" para saber qué operaciones puede hacer con ellos. No podemos sumar un "Grupo Sanguíneo" con una "Edad", ¿cierto?

Hoy conoceremos los 3 tipos más importantes para la bioestadística:
1. **Numeric (Numéricos):** Números con los que podemos hacer cálculos.
2. **Character (Caracteres/Texto):** Etiquetas de texto o categorías.
3. **Logical (Lógicos):** Valores de Verdadero o Falso.

## 1. Datos Numéricos (`numeric`)
Son los que ya hemos usado: peso, talla, frecuencia cardíaca o niveles de laboratorio. R permite decimales de forma automática.

**Instrucción:** Ejecuta la celda para guardar un valor de laboratorio.

```python
hemoglobina <- 14.2
# R reconoce que esto es un número y permite cálculos matemáticos
```

## 2. Datos de Texto (`character`)
Se usan para nombres, diagnósticos o categorías.
**Regla de Oro:** Todo lo que sea texto DEBE ir entre comillas `" "`. Si no pones comillas, R pensará que buscas una variable y marcará error.

**Instrucción:** Ejecuta la celda para guardar información de texto.

```python
paciente_nombre <- "Juan Pérez"
diagnostico <- "Hipertensión Arterial"
grupo_sanguineo <- "O+"
```

## 3. Datos Lógicos o Booleanos (`logical`)
Son muy útiles para marcar presencias o ausencias (Sí/No). Solo tienen dos valores posibles: `TRUE` (Verdadero) o `FALSE` (Falso). Deben escribirse siempre en mayúsculas y sin comillas.

**Instrucción:** Ejecuta la celda.

```python
fumador <- TRUE
alergias <- FALSE
```

## 4. ¿Cómo saber qué tipo de dato tengo?
A veces las bases de datos son grandes y no estamos seguros de cómo R leyó la información. Para eso usamos la función `class()`.

**Instrucción:** Ejecuta la celda y observa qué responde R.

```python
# Preguntamos por el tipo de dato de nuestras variables anteriores
class(hemoglobina)
class(diagnostico)
class(fumador)
```
