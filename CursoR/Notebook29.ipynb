# CS106 - Sesión 6 | Notebook 29: Relación entre variables (Scatter Plots) 📈

El gráfico de dispersión es la herramienta fundamental para explorar **correlaciones**. Cada punto representa a un paciente, ubicado según sus valores en dos ejes numéricos (X e Y).

Aprenderemos a:
1. Crear el gráfico básico de puntos.
2. Añadir una línea de tendencia.
3. Usar colores para ver cómo un tercer factor (como el sexo) influye en esa relación.

## 1. ¿Hay relación entre Peso y Estatura?
Vamos a ver si nuestra cohorte sigue la lógica física esperada.

**Instrucción:** Definiremos `estatura_m` en el eje X y `peso_kg` en el eje Y.

```r
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# Gráfico de dispersión básico
ggplot(datos, aes(x = estatura_m, y = peso_kg)) +
  geom_point(color = "darkblue", alpha = 0.5) +
  labs(title = "Relación entre Estatura y Peso",
       x = "Estatura (m)",
       y = "Peso (kg)") +
  theme_minimal()
```

## 2. Añadiendo la línea de "mejor ajuste"
Para ayudar al ojo a ver la tendencia, podemos añadir una capa llamada `geom_smooth()`. Por defecto, esto dibuja una curva suave que nos indica la dirección de los datos.

**Instrucción:** Ejecuta la celda y observa la zona sombreada (que representa el intervalo de confianza).

```r
ggplot(datos, aes(x = estatura_m, y = peso_kg)) +
  geom_point(alpha = 0.4) +
  geom_smooth(method = "lm", color = "red") + # "lm" para línea recta
  labs(title = "Tendencia Lineal: Estatura vs Peso",
       subtitle = "La línea roja indica la relación promedio",
       x = "Estatura (m)",
       y = "Peso (kg)")
```

## 3. Dispersión por categorías
¿La relación Edad vs Glucosa es igual para hombres y mujeres? Podemos colorear los puntos según una variable categórica.

**Instrucción:** Usaremos `color = sexo` dentro de la estética.

```r
ggplot(datos, aes(x = edad, y = glucosa_mgdl, color = sexo)) +
  geom_point(size = 2) +
  labs(title = "Glucosa vs Edad por Sexo",
       x = "Edad (años)",
       y = "Glucosa (mg/dL)",
       color = "Sexo")
```
