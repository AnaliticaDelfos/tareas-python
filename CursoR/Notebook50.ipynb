Has completado el recorrido por la estadística de supervivencia. En este reto final, no analizaremos la mortalidad, sino la **eficiencia hospitalaria**: el tiempo hasta el alta o la recuperación.

### El Escenario:
La dirección del hospital sospecha que ciertos grupos de pacientes tienen estancias significativamente más largas, lo que satura las camas. Tu misión es utilizar **Kaplan-Meier**, **Log-Rank** y **Cox** para identificar qué grupo "sobrevive" más tiempo dentro del hospital (es decir, quién tarda más en irse a casa).

## Instrucciones para el Investigador
Deberás ejecutar el análisis integral que compara el tiempo al evento (`dias_hasta_evento`) según el **Diagnóstico**.

**Tu informe debe responder:**
1. ¿Cuál es el tiempo mediano de estancia general?
2. ¿Existen diferencias estadísticas entre los diagnósticos?
3. ¿Cuál es la magnitud del riesgo (Hazard Ratio) de tener una estancia prolongada según la patología?

```python
# --- 1. CÓDIGO INTEGRAL DE ANÁLISIS ---
library(tidyverse)
library(survival)
library(survminer)

datos <- read_csv("cohorte_analisis.csv")

# A. Curva de Kaplan-Meier por Diagnóstico
ajuste_reto <- survfit(Surv(dias_hasta_evento, evento) ~ diagnostico, data = datos)

# Visualización
print(ggsurvplot(ajuste_reto, data = datos,
                 pval = TRUE,
                 risk.table = TRUE,
                 main = "Análisis de Estancia por Diagnóstico"))

# B. Prueba de Log-Rank
test_reto <- survdiff(Surv(dias_hasta_evento, evento) ~ diagnostico, data = datos)
print(test_reto)

# C. Modelo de Cox (Hazard Ratio)
modelo_reto <- coxph(Surv(dias_hasta_evento, evento) ~ diagnostico, data = datos)
print(summary(modelo_reto))

# --- 2. SECCIÓN DE RESPUESTAS (RELLENA LOS VALORES) ---

# r1: ¿Cuál es la mediana de supervivencia GLOBAL (días)?
# (Pista: mira el resultado de survfit(Surv(...) ~ 1))
r1_mediana_global <- 0

# r2: Según el Log-Rank, ¿hay diferencia significativa entre diagnósticos? (TRUE o FALSE)
r2_sig_logrank <- NA

# r3: Identifica el Hazard Ratio (exp(coef)) más alto en la tabla de Cox.
r3_hr_maximo <- 0.0000

# r4: Un Hazard Ratio de 2.0 en este contexto de "Alta Hospitalaria" significa:
# A) El paciente tiene el doble de riesgo de quedarse internado.
# B) El paciente tiene el doble de probabilidad de ser dado de alta rápido.
# Escribe "A" o "B"
r4_interpretacion_hr <- ""

# --- 3. CONSTRUCCIÓN DEL DATA FRAME DE EVALUACIÓN ---
df_respuesta <- data.frame(
  item = c("mediana_global", "diferencia_grupos", "hr_pico", "logica_clinica"),
  valor = c(as.character(r1_mediana_global),
            as.character(r2_sig_logrank),
            as.character(r3_hr_maximo),
            r4_interpretacion_hr)
)

print(df_respuesta)
```
