### Tu Misión:
La dirección médica necesita un informe epidemiológico de riesgos para entender qué patologías aceleran de forma crítica el ingreso hospitalario de los pacientes. Deberás aplicar de forma consecutiva las tres herramientas del módulo:
1. **Kaplan-Meier Global:** Para estimar el tiempo mediano general de la población libre de evento.
2. **Log-Rank Test:** Para evaluar si la presencia de Diabetes marca una brecha estadísticamente significativa.
3. **Modelo de Cox Multivariado:** Para meter todos los predictores a la licuadora (`diabetes`, `hipertension`, `obesidad`, `edad`, `sexo`) y aislar cuál es la patología más peligrosa.

```python
# --- 1. CÓDIGO INTEGRAL DE ANÁLISIS ---
library(tidyverse)
library(survival)

# Cargamos el archivo definitivo de la cohorte
datos <- read_csv("cohorte_survival.csv")

cat("=== PASO A: ESTIMACIÓN GLOBAL ===\n")
ajuste_global <- survfit(Surv(dias_hasta_hospitalizacion, hospitalizado) ~ 1, data = datos)
print(ajuste_global)

cat("\n=== PASO B: LOG-RANK TEST (DIABETES) ===\n")
test_reto <- survdiff(Surv(dias_hasta_hospitalizacion, hospitalizado) ~ diabetes, data = datos)
print(test_reto)

cat("\n=== PASO C: MODELO DE COX MULTIVARIADO DE MULTIMORBILIDAD ===\n")
# El modelo definitivo que evalúa todas las condiciones juntas
modelo_reto <- coxph(Surv(dias_hasta_hospitalizacion, hospitalizado) ~ diabetes + hipertension + obesidad + edad + sexo, data = datos)
print(summary(modelo_reto))


# --- 2. SECCIÓN DE RESPUESTAS (VALORES EXACTOS OBTENIDOS DE LA CONSOLA) ---

# r1: ¿Cuál es la mediana de tiempo libre de hospitalización GLOBAL (días)?
# (Míralo en el reporte del PASO A, columna 'median')
r1_mediana_global <- 93

# r2: Según el Log-Rank (PASO B), ¿existe una diferencia estadísticamente significativa
# en el tiempo a hospitalización entre diabéticos y no diabéticos? (Escribe TRUE o FALSE)
r2_sig_logrank <- TRUE

# r3: Identifica el Hazard Ratio (columna exp(coef)) MÁS ALTO entre las tres comorbilidades analizadas.
# (Revisa el PASO C y coloca el valor con 4 decimales)
r3_hr_maximo <- 2.4217

# r4: Un Hazard Ratio de 2.42 en la variable 'diabetes' dentro de este modelo multivariado significa:
# A) Los pacientes con diabetes tienen 2.42 veces más riesgo de hospitalizarse rápidamente en el tiempo.
# B) Los pacientes con diabetes tardan 2.42 veces más tiempo en requerir hospitalización.
# (Escribe únicamente "A" o "B")
r4_interpretacion_hr <- "A"


# --- 3. CONSTRUCCIÓN DEL DATA FRAME DE EVALUACIÓN ---
df_respuesta <- data.frame(
  item = c("mediana_global", "diferencia_grupos", "hr_pico", "logica_clinica"),
  valor = c(as.character(r1_mediana_global),
            as.character(r2_sig_logrank),
            as.character(r3_hr_maximo),
            r4_interpretacion_hr)
)

print(df_respuesta)
```
