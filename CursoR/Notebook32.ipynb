Una tabla descriptiva en un artículo científico debe ser "autoexplicativa". Esto significa que los nombres de las variables deben ser claros, incluir sus unidades de medida y tener un formato impecable.

En este cuaderno aprenderemos a:
1. **Traducir etiquetas:** Cambiar los nombres de las columnas del código a nombres clínicos.
2. **Controlar la estadística:** Decidir manualmente si queremos mostrar la Media $(\bar{x})$ y Desviación Estándar $(\sigma)$ o la Mediana $(\tilde{x})$ y el Rango Intercuartílico.
3. **Refinar el formato:** Aplicar negritas y controlar el número de decimales.

```python
library(tidyverse)
library(gtsummary)
datos <- read_csv("cohorte_analisis.csv")

datos |>
  select(edad, sexo, glucosa_mgdl, diagnostico) |>
  tbl_summary(
    label = list(
      edad ~ "Edad (años)",
      sexo ~ "Sexo biológico",
      glucosa_mgdl ~ "Glucosa sérica (mg/dL)",
      diagnostico ~ "Diagnóstico principal"
    )
  ) |>
  bold_labels() # Hace que los nombres de las variables resalten en negrita
```

## 2. Media vs. Mediana
`gtsummary` es conservador y usa la Mediana por defecto para números. Sin embargo, si tus datos son normales (como la estatura), podrías preferir la Media.

Podemos usar el argumento `statistic` para definir qué queremos ver:
* `{mean} ({sd})` para Media y Desviación Estándar.
* `{median} ({p25}, {p75})` para Mediana y Percentiles (IQR).

```python
datos |>
  select(edad, estatura_m, peso_kg) |>
  tbl_summary(
    label = list(
      edad ~ "Edad",
      estatura_m ~ "Estatura (m)",
      peso_kg ~ "Peso (kg)"
    ),
    statistic = list(
      # Queremos media para estatura y peso, pero mediana para edad
      all_continuous() ~ "{mean} ({sd})",
      edad ~ "{median} ({p25}, {p75})"
    ),
    digits = all_continuous() ~ 1 # Forzamos a que solo muestre 1 decimal
  )
```

## 3. Tipografía y Encabezados Profesionales
Para que una tabla sea legible, debemos guiar el ojo del lector. `gtsummary` permite resaltar las variables y cambiar los títulos de las columnas que vienen por defecto en inglés.

Usaremos:
* **`bold_labels()`**: Pone los nombres de las variables en negrita.
* **`italicize_levels()`**: Pone las subcategorías (ej. M/F) en cursiva.
* **`modify_header()`**: Cambia el título de la columna "Characteristic" por algo más profesional como "**Variable Clínica**".

```python
datos |>
  select(edad, sexo, diagnostico) |>
  tbl_summary(
    label = list(edad ~ "Edad", sexo ~ "Sexo", diagnostico ~ "Diagnóstico")
  ) |>
  bold_labels() |>
  italicize_levels() |>
  modify_header(label = "**Variable Clínica**") # Nota los asteriscos para negrita en Markdown
```
