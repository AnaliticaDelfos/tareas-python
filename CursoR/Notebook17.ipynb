# CS106 - Sesión 4 | Notebook 17: Categorizando el riesgo (if_else / case_when) 🏷️

En la práctica clínica, los números por sí solos a veces necesitan una interpretación. No es lo mismo ver un "30" que leer "Obesidad".

Hoy aprenderemos a usar funciones lógicas dentro de `mutate()` para asignar etiquetas automáticas a nuestros pacientes:
* **`if_else()`**: Para decisiones sencillas (Sí/No, Normal/Anormal).
* **`case_when()`**: Para clasificaciones complejas con múltiples niveles (ej. Rangos de IMC o estadios de una enfermedad).

## 1. Clasificación simple con `if_else()`
La función funciona así: `if_else(condición, "qué pasa si es cierto", "qué pasa si es falso")`.

**Instrucción:** Vamos a crear una columna que nos diga si el paciente es "Adulto Mayor" (60 años o más) o "Adulto".

```python
library(tidyverse)
datos <- read_csv("cohorte_estudio.csv")

# Clasificamos por edad
datos_etiquetados <- datos |>
  mutate(grupo_etario = if_else(edad >= 60, "Adulto Mayor", "Adulto"))

# Revisamos el resultado
datos_etiquetados |> select(nombre, edad, grupo_etario)
```

## 2. Clasificaciones complejas con `case_when()`
Cuando tenemos más de dos opciones, `case_when()` es mucho más ordenado. Se usa el símbolo `~` (virgulilla) para decir "entonces asigna esto".

**Instrucción:** Clasifiquemos los niveles de glucosa en tres categorías:
* Menos de 100: "Normal"
* Entre 100 y 125: "Pre-diabetes"
* 126 o más: "Diabetes"

```python
datos_diagnostico <- datos |>
  mutate(estado_glucosa = case_when(
    glucosa_mgdl < 100  ~ "Normal",
    glucosa_mgdl <= 125 ~ "Pre-diabetes",
    glucosa_mgdl >= 126 ~ "Diabetes",
    TRUE                ~ "Sin dato" # Para cualquier otro caso
  ))

datos_diagnostico |> select(nombre, glucosa_mgdl, estado_glucosa)
```
