Un modelo predictivo puede tener valores p muy bajos y, sin embargo, ser un pésimo predictor. Para evaluar la "bondad de ajuste", utilizamos el **$R^2$ (R-cuadrado)**.

### El Coeficiente de Determinación ($R^2$):
Representa la proporción de la variabilidad del desenlace que es explicada por el modelo.
* **$R^2 = 0.80$**: El modelo explica el 80% de lo que le sucede al paciente. El otro 20% se debe a factores no medidos o azar.
* **$R^2$ Ajustado:** Es una versión más castigada y realista que usamos en modelos con múltiples predictores para evitar el sobreajuste.



### Los Residuos:
Son la diferencia entre el valor real del paciente y lo que el modelo predijo. Si los residuos son grandes o tienen patrones, nuestro modelo está fallando en algo importante.

## 1. ¿Cuánto explicamos de la realidad?
Retomaremos el modelo donde predecimos el **Peso** usando la **Estatura**. Veremos qué tanto de la variación del peso corporal se debe únicamente a la talla del paciente.

```r
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# Generamos el modelo simple
modelo_eval <- lm(peso_kg ~ estatura_m, data = datos)

# Extraemos el resumen
resumen_stats <- summary(modelo_eval)

# Mostramos específicamente el R-cuadrado
cat("R-cuadrado (Original):", resumen_stats$r.squared, "\n")
cat("R-cuadrado Ajustado:", resumen_stats$adj.r.squared)
```

## 2. Diagnóstico del Modelo
Un buen modelo debe tener residuos distribuidos al azar (como una nube de puntos sin forma). Si vemos curvas o tendencias en los residuos, significa que la relación quizás no es lineal o nos faltan variables.

```r
# Graficamos los residuos vs los valores ajustados
plot(modelo_eval, which = 1, col = "steelblue", pch = 16)
```
