# CS106 - Sesión 6 | Notebook 27: Boxplots y Outliers 📦

¿Cómo sabemos si un paciente tiene una glucosa "normalmente alta" o si es un error de captura o un caso extremo? El **Boxplot** es la herramienta estándar para esto.

Un Boxplot nos muestra:
1.  **La Mediana:** La línea central.
2.  **Caja (Rango Intercuartílico):** Donde está el 50% central de los pacientes.
3.  **Bigotes:** El rango esperado de los datos.
4.  **Puntos (Outliers):** Valores que están fuera de lo esperado.

## 1. Construyendo una caja
A diferencia del histograma, el boxplot suele llevar la variable numérica en el eje **Y**.

**Instrucción:** Vamos a graficar la distribución del Peso en toda la cohorte.

```python
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# Graficamos el Peso
ggplot(datos, aes(y = peso_kg)) +
  geom_boxplot(fill = "lightgreen") +
  labs(title = "Distribución de Peso Corporal",
       y = "Peso (kg)")
```

## 2. Comparando Grupos (La verdadera potencia)
El boxplot brilla cuando comparamos categorías. ¿Tienen los hombres más glucosa que las mujeres? ¿Los pacientes hospitalizados son más añosos?

**Instrucción:** Colocaremos una variable categórica en el eje **X** y la numérica en el eje **Y**.

```python
ggplot(datos, aes(x = sexo, y = glucosa_mgdl, fill = sexo)) +
  geom_boxplot() +
  labs(title = "Glucosa por Sexo",
       x = "Sexo",
       y = "Glucosa (mg/dL)") +
  theme_minimal()
```
