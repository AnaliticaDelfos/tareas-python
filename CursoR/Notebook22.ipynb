# CS106 - Sesión 5 | Notebook 22: Imputación por Medidas de Tendencia Central 📈

Cuando faltan datos numéricos (como edad, peso o niveles de glucosa), borrar al paciente completo es un desperdicio de información. Una técnica común es la **imputación simple**, que consiste en llenar los huecos (`NA`) con:
* **La Media (Promedio):** Ideal si los datos están bien distribuidos.
* **La Mediana:** Ideal si hay valores muy extremos (outliers) que afectan el promedio.

Hoy aprenderemos a usar `mutate()` para realizar este reemplazo de forma masiva en nuestra cohorte de 200 pacientes.

## 1. ¿Qué valor vamos a usar?
Antes de llenar los huecos, debemos calcular el valor que usaremos como "parche". Es vital usar el argumento `na.rm = TRUE` para que R ignore los vacíos actuales al calcular el promedio.

**Instrucción:** Calculemos la mediana de la edad en nuestra base de datos.

```r
library(tidyverse)
datos <- read_csv("cohorte_limpieza.csv")

# Calculamos la mediana de edad
mediana_edad <- median(datos$edad, na.rm = TRUE)

# Mostramos el valor que usaremos para imputar
mediana_edad
```

## 2. Llenando los huecos con `replace_na()`
La función `replace_na()` es la forma más directa de sustituir los `NA`. Le pasamos una lista con el nombre de la columna y el valor que queremos poner.

**Instrucción:** Vamos a crear una nueva base donde la columna `edad` ya no tenga datos faltantes.

```r
# Aplicamos la imputación
datos_imputados <- datos |>
  mutate(edad = replace_na(edad, mediana_edad))

# Verificamos si aún quedan NAs en edad
sum(is.na(datos_imputados$edad))
```

## 3. El método `if_else()`
A veces queremos más control. Podemos usar `if_else()` para decir: "Si el dato es NA, pon la mediana; si no es NA, deja el dato original".

**Instrucción:** Imputemos la glucosa usando este método.

```r
promedio_glucosa <- mean(datos$glucosa_mgdl, na.rm = TRUE)

datos |>
  mutate(glucosa_mgdl = if_else(is.na(glucosa_mgdl), promedio_glucosa, glucosa_mgdl)) |>
  select(nombre, glucosa_mgdl)
```
