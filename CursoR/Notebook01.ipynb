### ¿Qué es un Notebook?
Es un documento compuesto por "celdas". Principalmente usaremos dos tipos:
1. **Celdas de Texto (Markdown):** Como esta que estás leyendo, donde anotamos explicaciones o notas clínicas.
2. **Celdas de Código:** Bloques donde escribiremos instrucciones que el lenguaje **R** ejecutará.

## 1. ¿Cómo "correr" el código?
Para que R procese una instrucción, debes seleccionar la celda de código y presionar:
> **`Shift + Enter`** (o hacer clic en el botón **Run** en la barra superior).

Cuando una celda se está ejecutando, verás un asterisco `In [*]`. Cuando termina, aparecerá un número `In [1]`.
