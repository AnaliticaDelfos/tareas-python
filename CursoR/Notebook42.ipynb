Si la correlación nos decía que dos variables "se llevan", la regresión nos dice exactamente **cuánto se afectan**. En medicina, esto nos permite crear modelos predictivos.

### La Ecuación de la Recta:
$$Y = \beta_0 + \beta_1 X + \epsilon$$

* **$Y$ (Variable Dependiente):** El desenlace que queremos predecir (ej. Peso).
* **$X$ (Variable Independiente):** El factor que explica el cambio (ej. Estatura).
* **$\beta_0$ (Intercepto):** El valor de $Y$ cuando $X$ es cero.
* **$\beta_1$ (Pendiente/Coeficiente):** Es el valor más importante. Indica cuánto cambia $Y$ por cada unidad que aumenta $X$.

## 1. Ajustando la Línea de Regresión
Utilizaremos la función `lm()` (Linear Model). El objetivo es encontrar la línea que minimice la distancia entre los puntos reales de los pacientes y nuestras predicciones.

```python
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# Creamos el modelo: El Peso depende de la Estatura
modelo_simple <- lm(peso_kg ~ estatura_m, data = datos)

# Visualizamos la línea sobre los datos
ggplot(datos, aes(x = estatura_m, y = peso_kg)) +
  geom_point(alpha = 0.4) +
  geom_smooth(method = "lm", color = "darkgreen") +
  labs(title = "Modelo de Predicción de Peso",
       x = "Estatura (m)", y = "Peso (kg)") +
  theme_light()
```

## 2. ¿Cómo leer el reporte de regresión?
Al ejecutar el `summary()`, debemos fijarnos en la columna **Estimate**.
* El valor asociado a la variable (estatura_m) nos dice cuántos kilos extra tiene un paciente por cada metro adicional de estatura.

```python
# Mostramos el reporte detallado del modelo
summary(modelo_simple)
```
