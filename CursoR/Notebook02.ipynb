## 2. Conversión de Unidades
Muchas veces recibimos datos en unidades que necesitamos convertir. Por ejemplo, pasar la temperatura de grados **Fahrenheit** a **Celsius**.

La fórmula es: `(Fahrenheit - 32) / 1.8`

**Ejercicio:** Un paciente llega con 101°F de fiebre. ¿Cuánto es en Celsius? Ejecuta la celda:

```python
# Conversión de 101°F a Celsius
(101 - 32) / 1.8
```

## 3. El Índice de Masa Corporal (IMC)
El IMC es una medida estándar que se calcula dividiendo el peso (kg) entre la estatura al cuadrado ($m^2$).

La fórmula es: `peso / (estatura * estatura)`

**Ejercicio:** Calcula el IMC de una persona que pesa 85 kg y mide 1.72 metros.

```python
# Cálculo de IMC directo
85 / (1.72 * 1.72)
```
