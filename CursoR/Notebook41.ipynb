# CS106 - Sesión 9 | Notebook 41: Correlación (Pearson vs. Spearman) 📈

La correlación cuantifica la fuerza y la dirección de la relación entre dos variables numéricas. Es el primer paso para entender el riesgo: ¿si aumenta un factor, aumenta también el desenlace?

### Interpretación del Coeficiente (r):
* **Magnitud:** Va de 0 (sin relación) a 1 (relación perfecta).
* **Dirección:**
    * **Positiva (+):** Ambas variables aumentan juntas (ej. Edad y Presión Arterial).
    * **Negativa (-):** Una aumenta y la otra disminuye (ej. Ejercicio y Frecuencia Cardíaca en reposo).



[Image of scatter plots showing positive correlation, negative correlation, and no correlation]


### ¿Qué prueba elegir?
1. **Pearson:** Para variables con distribución **Normal**.
2. **Spearman:** Para variables **No Normales** o cuando hay valores extremos (outliers).

## 1. Inspección Visual: Scatter Plot
Antes de calcular números, debemos "ver" la relación. Un gráfico de dispersión nos permite identificar si la tendencia es lineal o si hay pacientes que se comportan de forma muy atípica.

```python
library(tidyverse)
datos <- read_csv("cohorte_analisis.csv")

# Gráfico de dispersión con línea de tendencia (regresión)
ggplot(datos, aes(x = estatura_m, y = peso_kg)) +
  geom_point(alpha = 0.5, color = "steelblue") +
  geom_smooth(method = "lm", color = "red", se = TRUE) +
  labs(title = "Relación entre Estatura y Peso",
       subtitle = "La línea roja indica la tendencia promedio",
       x = "Estatura (m)", y = "Peso (kg)") +
  theme_minimal()
```

## 2. El Veredicto Matemático
Calcularemos los coeficientes para validar nuestra observación visual. Reportaremos el valor **p** para saber si la relación es estadísticamente significativa.

```python
# Correlación de Pearson (Asumiendo normalidad)
res_pearson <- cor.test(datos$estatura_m, datos$peso_kg, method = "pearson")
print(res_pearson)

# Correlación de Spearman (No paramétrica / Rangos)
res_spearman <- cor.test(datos$estatura_m, datos$peso_kg, method = "spearman")
print(res_spearman)
```
