### Las herramientas: Librerías (Paquetes)
R por sí solo es muy potente, pero a veces necesita "herramientas especiales" para tareas específicas, como leer archivos modernos. Estas herramientas se llaman **paquetes** o **librerías**.

Usaremos dos principales:
1. **`readr`**: Especializada en archivos CSV (Comma Separated Values).
2. **`readxl`**: Especializada en archivos de Microsoft Excel (.xlsx).

## 1. Activando nuestras herramientas
Para usar un paquete, primero debemos "llamarlo" usando la función `library()`. Es como abrir la caja de herramientas antes de empezar a trabajar.

**Instrucción:** Ejecuta la celda de abajo para activar las librerías.

```python
# Activamos las librerías necesarias
library(readr)
library(readxl)
```

## 2. Leyendo archivos CSV

**Instrucción:** Vamos a cargar una pequeña base de datos de ejemplo llamada `cohorte_estudio.csv` que ya está en tu carpeta de trabajo.

```python
# Cargamos el archivo y lo guardamos en una variable llamada 'datos'
datos <- read_csv("cohorte_estudio.csv")

# Vemos una vista previa de lo que importamos
datos
```
