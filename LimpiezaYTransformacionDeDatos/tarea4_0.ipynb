```python
@Controlador("4", "ejercicio_lytd_16")
def ejercicio_4_16(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame que contiene
        las columnas:

        nombre, edad, salario

        Debes usar loc para obtener las filas donde:
        - edad sea mayor a 30
        - salario sea mayor a 20000

    Ejemplo:

        nombre edad salario
        Ana     35   25000
        Luis    28   30000
        Pedro   40   18000

        Resultado esperado:

        nombre edad salario
        Ana     35   25000

    Recuerda:
        Para combinar condiciones usa el operador &

    Recuerda:
        Cada condición debe ir entre paréntesis.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.loc[(df["edad"] > 30) & (df["salario"] > 20000)]

    # Código respuesta

    return respuesta
```

```python
@Controlador("4", "ejercicio_lytd_17")
def ejercicio_4_17(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame que contiene
        las columnas:

        nombre, edad, ciudad

        Debes usar loc para obtener las filas donde:
        - la ciudad sea "CDMX"
        - la edad sea menor a 30

    Ejemplo:

        nombre edad ciudad
        Ana     25  CDMX
        Luis    40  CDMX
        Pedro   22  Monterrey

        Resultado esperado:

        nombre edad ciudad
        Ana     25  CDMX

    Recuerda:
        Puedes combinar condiciones usando &

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.loc[(df["ciudad"] == "CDMX") & (df["edad"] < 30)]

    # Código respuesta

    return respuesta
```

```python
@Controlador("4", "ejercicio_lytd_18")
def ejercicio_4_18(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con las columnas:

        producto, precio, stock

        Debes usar loc para obtener las filas donde:
        - el precio sea mayor a 500
        - el stock sea menor a 10

    Ejemplo:

        producto precio stock
        Laptop   20000  5
        Mouse     300   50
        Monitor  4000   8

        Resultado esperado:

        producto precio stock
        Laptop   20000  5
        Monitor  4000   8

    Recuerda:
        Usa & para combinar condiciones.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.loc[(df["precio"] > 500) & (df["stock"] < 10)]

    # Código respuesta

    return respuesta
```

```python
@Controlador("4", "ejercicio_lytd_19")
def ejercicio_4_19(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con las columnas:

        nombre, edad, salario

        Debes usar loc para obtener las filas donde:
        - edad sea menor a 25
        O
        - salario sea mayor a 50000

    Ejemplo:

        nombre edad salario
        Ana     22   20000
        Luis    35   60000
        Pedro   40   30000

        Resultado esperado:

        nombre edad salario
        Ana     22   20000
        Luis    35   60000

    Recuerda:
        Para condiciones OR se usa el operador |

    Recuerda:
        Cada condición debe ir entre paréntesis.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.loc[(df["edad"] < 25) | (df["salario"] > 50000)]

    # Código respuesta

    return respuesta
```

```python
ejercicio_4_19([[1], [2], [3]], [2, 4, 5])
```

```python
calificador.califica_ejercicio_lytd_19(ejercicio_4_19)
```

```python
@Controlador("4", "ejercicio_lytd_20")
def ejercicio_4_20(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con las columnas:

        producto, precio

        Debes usar loc para modificar el DataFrame
        de forma que:

        - todos los productos con precio menor a 100
        - o mayor a 10000

        tengan el valor de precio igual a 100.

    Ejemplo:

        producto precio
        Mouse     50
        Laptop   20000
        Teclado   800

        Resultado esperado:

        producto precio
        Mouse    100
        Laptop   100
        Teclado  800

    Recuerda:
        Puedes modificar valores usando loc.

    Recuerda:
        Usa | para combinar condiciones.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df.loc[(df["precio"] < 100) | (df["precio"] > 10000), "precio"] = 100

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_4_20([[1, 2], [2, 3], [3, 4]], [3, 6, 9], [[4, 5], [5, 6]])
```

```python
calificador.califica_ejercicio_lytd_20(ejercicio_4_20)
```
