```python
@Controlador("2", "ejercicio_lytd_6")
def ejercicio_2_6(lista_productos: list, lista_precios: list):
    """
    Instrucciones:
        Esta función recibe dos listas:
        - lista_productos
        - lista_precios

        Debes crear una Serie de pandas donde:
        - los índices sean los nombres de los productos
        - los valores sean los precios

    Ejemplo:
        lista_productos = ["Laptop", "Mouse", "Teclado"]
        lista_precios = [20000, 300, 800]

        Resultado esperado:

        Laptop     20000
        Mouse        300
        Teclado      800

    Recuerda:
        En una Serie los índices pueden ser etiquetas.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = pd.Series(lista_precios, index=lista_productos)

    # Código respuesta

    return respuesta
```

```python
@Controlador("2", "ejercicio_lytd_7")
def ejercicio_2_7(lista_nombres: list, lista_calif1: list, lista_calif2: list):
    """
    Instrucciones:
        Esta función recibe tres listas:
        - lista_nombres
        - lista_calif1
        - lista_calif2

        Debes crear un DataFrame con tres columnas:
        "nombre", "calif1", "calif2".

        Después debes crear una nueva columna llamada
        "promedio" que sea el promedio de calif1 y calif2.

    Ejemplo:
        nombre  calif1  calif2  promedio
        Ana       8       10       9
        Luis      7        9       8

    Recuerda:
        Las columnas de un DataFrame se pueden operar
        entre sí directamente.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    datos = {
        "nombre": lista_nombres,
        "calif1": lista_calif1,
        "calif2": lista_calif2
    }

    df = pd.DataFrame(datos)

    df["promedio"] = (df["calif1"] + df["calif2"]) / 2

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_2_7(["Ana", "Luis", "Pepe"], [10, 9, 8], [10, 7, 10])
```

```python
calificador.califica_ejercicio_lytd_7(ejercicio_2_7)
```

```python
@Controlador("1", "ejercicio_lytd_8")
def ejercicio_2_8(lista_paises: list, lista_capitales: list, lista_poblacion: list):
    """
    Instrucciones:
        Esta función recibe tres listas:
        - lista_paises
        - lista_capitales
        - lista_poblacion

        Debes crear un DataFrame donde:
        - las columnas sean "capital" y "poblacion"
        - los índices sean los nombres de los países

    Ejemplo:

        capital      poblacion
    Mexico  CDMX       9000000
    Francia Paris      2100000

    Recuerda:
        Puedes asignar el índice usando el parámetro
        index al crear el DataFrame.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    datos = {
        "capital": lista_capitales,
        "poblacion": lista_poblacion
    }

    respuesta = pd.DataFrame(datos, index=lista_paises)

    # Código respuesta

    return respuesta
```

```python
@Controlador("1", "ejercicio_lytd_10")
def ejercicio_2_10(lista_nombres: list, lista_ventas: list):
    """
    Instrucciones:
        Esta función recibe dos listas:
        - lista_nombres
        - lista_ventas

        Debes crear un DataFrame con dos columnas:
        "vendedor" y "ventas".

        Después debes agregar una tercera columna llamada
        "ventas_miles" que contenga las ventas divididas entre 1000.

    Ejemplo:

        vendedor  ventas  ventas_miles
        Ana       5000        5
        Luis      12000      12

    Recuerda:
        Puedes crear nuevas columnas realizando
        operaciones matemáticas sobre columnas existentes.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    datos = {
        "vendedor": lista_nombres,
        "ventas": lista_ventas
    }

    df = pd.DataFrame(datos)

    df["ventas_miles"] = df["ventas"] / 1000

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_2_10(["Ana", "Luis"], [5000, 12000])
```

```python
calificador.califica_ejercicio_lytd_10(ejercicio_2_10)
```
