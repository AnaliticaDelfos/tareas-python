```python
@Controlador("6", "ejercicio_lytd_26")
def ejercicio_6_26(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "genero" que contiene los valores:

        "M" y "F"

        Debes reemplazar:
        "M" por "Masculino"
        "F" por "Femenino"

        usando replace().

    Ejemplo:

        genero
        M
        F
        M

        Resultado esperado:

        genero
        Masculino
        Femenino
        Masculino

    Recuerda:
        replace puede usar diccionarios.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df["genero"] = df["genero"].replace({
        "M": "Masculino",
        "F": "Femenino"
    })

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
@Controlador("6", "ejercicio_lytd_27")
def ejercicio_6_27(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "pais" que contiene códigos de país:

        "MX", "US", "ES"

        Debes crear una nueva columna llamada
        "pais_nombre" usando map() y el siguiente diccionario:

        {
            "MX": "Mexico",
            "US": "Estados Unidos",
            "ES": "España"
        }

    Ejemplo:

        pais
        MX
        US

        Resultado esperado:

        pais pais_nombre
        MX   Mexico
        US   Estados Unidos

    Recuerda:
        map permite transformar valores usando diccionarios.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    mapa = {
        "MX": "Mexico",
        "US": "Estados Unidos",
        "ES": "España"
    }

    df["pais_nombre"] = df["pais"].map(mapa)

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_6_27(["Ana", "Luis"], [20, 25])
```

```python
calificador.califica_ejercicio_lytd_27(ejercicio_6_27)
```

```python
@Controlador("6", "ejercicio_lytd_28")
def ejercicio_6_28(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "estatus" que contiene valores numéricos:

        0 y 1

        Debes reemplazar:
        0 por "Inactivo"
        1 por "Activo"

        usando replace().

    Ejemplo:

        estatus
        1
        0
        1

        Resultado esperado:

        estatus
        Activo
        Inactivo
        Activo

    Recuerda:
        replace también funciona con valores numéricos.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df["estatus"] = df["estatus"].replace({
        0: "Inactivo",
        1: "Activo"
    })

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
@Controlador("6", "ejercicio_lytd_29")
def ejercicio_6_29(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "calificacion" con valores de 0 a 10.

        Debes crear una nueva columna llamada "resultado"
        usando map() y el siguiente diccionario:

        {
            10: "Excelente",
            9: "Excelente",
            8: "Bueno",
            7: "Bueno",
            6: "Aprobado"
        }

        Si algún valor no aparece en el diccionario,
        el resultado será NaN (esto es normal).

    Ejemplo:

        calificacion
        10
        8
        6

        Resultado esperado:

        calificacion resultado
        10          Excelente
        8           Bueno
        6           Aprobado

    Recuerda:
        map aplica transformaciones basadas en un diccionario.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    mapa = {
        10: "Excelente",
        9: "Excelente",
        8: "Bueno",
        7: "Bueno",
        6: "Aprobado"
    }

    df["resultado"] = df["calificacion"].map(mapa)

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_6_29([[1], [2], [3]], [2, 4, 5])
```

```python
calificador.califica_ejercicio_lytd_29(ejercicio_6_29)
```

```python
@Controlador("6", "ejercicio_lytd_30")
def ejercicio_6_30(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "categoria".

        Debes reemplazar los valores:

        "A" → "Alta"
        "B" → "Media"
        "C" → "Baja"

        usando replace().

    Ejemplo:

        categoria
        A
        C
        B

        Resultado esperado:

        categoria
        Alta
        Baja
        Media

    Recuerda:
        replace puede aplicar múltiples reemplazos.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df["categoria"] = df["categoria"].replace({
        "A": "Alta",
        "B": "Media",
        "C": "Baja"
    })

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_6_30([[1, 2], [2, 3], [3, 4]], [3, 6, 9], [[4, 5], [5, 6]])
```

```python
calificador.califica_ejercicio_lytd_30(ejercicio_6_30)
```
