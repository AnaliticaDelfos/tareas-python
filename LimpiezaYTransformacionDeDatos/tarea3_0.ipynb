```python
@Controlador("3", "ejercicio_lytd_11")
def ejercicio_3_11(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna llamada "edad".

        Debes usar loc para obtener únicamente las filas donde
        la edad sea mayor a 30.

    Ejemplo:
        Si el DataFrame es:

        nombre  edad
        Ana      25
        Luis     40
        Pedro    35

        Resultado esperado:

        nombre  edad
        Luis     40
        Pedro    35

    Recuerda:
        loc permite seleccionar filas usando condiciones.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.loc[df["edad"] > 30]

    # Código respuesta

    return respuesta
```

```python
@Controlador("3", "ejercicio_lytd_12")
def ejercicio_3_12(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame.

        Debes usar iloc para obtener las primeras
        tres filas del DataFrame.

    Ejemplo:
        Si el DataFrame tiene 10 filas,
        el resultado debe contener únicamente las filas
        en posición 0, 1 y 2.

    Recuerda:
        iloc selecciona filas usando posiciones.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.iloc[0:3]

    # Código respuesta

    return respuesta
```

```python
ejercicio_3_12(["Ana", "Luis"], [20, 25])
```

```python
@Controlador("3", "ejercicio_lytd_13")
def ejercicio_3_13(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame que contiene
        las columnas:

        nombre, edad, salario

        Debes usar loc para seleccionar únicamente
        las columnas "nombre" y "salario".

    Ejemplo:

        nombre salario
        Ana    20000
        Luis   25000

    Recuerda:
        loc permite seleccionar columnas usando
        una lista de nombres.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.loc[:, ["nombre", "salario"]]

    # Código respuesta

    return respuesta
```

```python
ejercicio_3_13([100, 200, 300], ["a", "b", "c"])
```

```python
calificador.califica_ejercicio_lytd_13(ejercicio_3_13)
```

```python
@Controlador("3", "ejercicio_lytd_14")
def ejercicio_3_14(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame.

        Debes usar iloc para obtener el valor que se
        encuentra en la segunda fila y tercera columna.

        Recuerda que iloc usa índices que empiezan en 0.

    Ejemplo:
        Si el DataFrame es:

            A   B   C
        0   1   2   3
        1   4   5   6

        El resultado debe ser:

        6

    Recuerda:
        iloc[fila, columna]

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.iloc[1, 2]

    # Código respuesta

    return respuesta
```

```python
ejercicio_3_14([[1], [2], [3]], [2, 4, 5])
```

```python
calificador.califica_ejercicio_lytd_14(ejercicio_3_14)
```

```python
@Controlador("3", "ejercicio_lytd_15")
def ejercicio_3_15(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame que contiene
        una columna llamada "ventas".

        Debes usar loc para modificar el DataFrame
        de modo que todas las filas donde ventas sea
        menor a 100 se conviertan en 100.

    Ejemplo:

        ventas
        50
        120
        80

        Resultado esperado:

        ventas
        100
        120
        100

    Recuerda:
        loc también se puede usar para modificar valores.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df.loc[df["ventas"] < 100, "ventas"] = 100

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_3_15([[1, 2], [2, 3], [3, 4]], [3, 6, 9], [[4, 5], [5, 6]])
```

```python
calificador.califica_ejercicio_lytd_15(ejercicio_3_15)
```
