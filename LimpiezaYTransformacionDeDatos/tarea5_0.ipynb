```python
@Controlador("5", "ejercicio_lytd_21")
def ejercicio_5_21(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame que contiene
        una columna llamada "edad".

        Debes usar isna() y loc para obtener únicamente
        las filas donde la edad sea un valor faltante (NaN).

    Ejemplo:

        nombre edad
        Ana    25
        Luis   NaN
        Pedro  30

        Resultado esperado:

        nombre edad
        Luis   NaN

    Recuerda:
        isna() permite detectar valores faltantes.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    respuesta = df.loc[df["edad"].isna()]

    # Código respuesta

    return respuesta
```

```python
@Controlador("5", "ejercicio_lytd_22")
def ejercicio_5_22(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "salario".

        Debes reemplazar todos los valores faltantes
        en la columna salario por 0 usando fillna().

    Ejemplo:

        salario
        20000
        NaN
        15000

        Resultado esperado:

        salario
        20000
        0
        15000

    Recuerda:
        fillna() permite reemplazar valores faltantes.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df["salario"] = df["salario"].fillna(0)

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_5_22(["Ana", "Luis"], [20, 25])
```

```python
@Controlador("5", "ejercicio_lytd_23")
def ejercicio_5_23(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "precio".

        Debes reemplazar los valores faltantes en esa
        columna con el promedio de la columna.

    Ejemplo:

        precio
        100
        NaN
        300

        El promedio de los valores existentes es 200.

        Resultado esperado:

        precio
        100
        200
        300

    Recuerda:
        Puedes usar mean() para calcular el promedio.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    promedio = df["precio"].mean()

    df["precio"] = df["precio"].fillna(promedio)

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_5_23([100, 200, 300], ["a", "b", "c"])
```

```python
calificador.califica_ejercicio_lytd_23(ejercicio_5_23)
```

```python
@Controlador("5", "ejercicio_lytd_24")
def ejercicio_5_24(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con dos columnas:

        nombre, ciudad

        Debes reemplazar todos los valores faltantes
        de la columna "ciudad" con el texto "Desconocido".

    Ejemplo:

        nombre ciudad
        Ana    CDMX
        Luis   NaN
        Pedro  Guadalajara

        Resultado esperado:

        nombre ciudad
        Ana    CDMX
        Luis   Desconocido
        Pedro  Guadalajara

    Recuerda:
        fillna también puede usar texto.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df["ciudad"] = df["ciudad"].fillna("Desconocido")

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
@Controlador("5", "ejercicio_lytd_25")
def ejercicio_5_25(df: pd.DataFrame):
    """
    Instrucciones:
        Esta función recibe un DataFrame con una columna
        llamada "ventas".

        Debes hacer lo siguiente:

        1) Encontrar las filas donde ventas es NaN usando isna()
        2) Reemplazar esos valores con 0 usando loc.

    Ejemplo:

        ventas
        500
        NaN
        300

        Resultado esperado:

        ventas
        500
        0
        300

    Recuerda:
        loc permite modificar filas que cumplen
        una condición.

    Recuerda:
        Cuida la indentación.
    """

    # Código respuesta

    df.loc[df["ventas"].isna(), "ventas"] = 0

    respuesta = df

    # Código respuesta

    return respuesta
```

```python
ejercicio_5_25([[1, 2], [2, 3], [3, 4]], [3, 6, 9], [[4, 5], [5, 6]])
```

```python
calificador.califica_ejercicio_lytd_25(ejercicio_5_25)
```
